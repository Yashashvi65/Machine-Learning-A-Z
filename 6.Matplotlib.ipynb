```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Graph") 
# plt.savefig("Graph.png") To save the figure locally
```

```python
y=x*x
```

```python
plt.plot(x,y,"ro")
```

```python
plt.plot(x,y,"ro",linestyle='dashed',linewidth=2,markersize=10)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"r--")
plt.subplot(2,2,2)
plt.plot(x,y,"b-o")
plt.subplot(2,2,3)
plt.plot(x,y,"g-x")
plt.subplot(2,2,4)
plt.plot(x,y,"r-v")
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
#Bar plot
```

```python
x=[5,6,7]
y=[3,6,9]
plt.bar(x,y,align='center')
```

```python
x=np.array([6,7,78,77,4,3,3,7])
plt.hist(x)
```

```python
 #Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```

```python
#Box Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);
```
